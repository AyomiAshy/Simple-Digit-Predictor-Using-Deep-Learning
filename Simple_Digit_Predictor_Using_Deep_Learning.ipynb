```python
import tensorflow as tf # Building the neural network
from tensorflow.keras import layers,models
import matplotlib.pyplot as plt
```

```python
# Load the data
(x_train,y_train),(x_test,y_test)=tf.keras.datasets.mnist.load_data()
```

```python
# Normalize the data
x_train=x_train/255.0
x_test=x_test/255.0
```

```python
# Build the Model
model=models.Sequential([
    # First Layer
    layers.Flatten(input_shape=(28,28)), # 28 by 28 pixels and converted into 1d
    # Dense Layer-Work on the data
    layers.Dense(128,activation='relu'),
    # Output Layer-Predictions
    layers.Dense(10,activation='softmax')
])
```

```python
# Compile the model
model.compile(
    optimizer='adam', # Adjust the weights efficiently
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'] # Evaluate the model's performance
)
```

```python
# Train the model
model.fit(
    x_train,y_train,
    epochs=50,
    validation_data=(x_test,y_test)
)
```

```python
# Evaluate the model
test_loss, test_acc=model.evaluate(x_test,y_test)
print(f"Model Accuracy: {test_acc}")
```

```python
# Make Predictions
predictions=model.predict(x_test)
```

```python
# Display the predicted digit
plt.imshow(x_test[0],cmap='gray') # Display the first testing digit
plt.title(predictions[0].argmax()) # Display the first digit to be predicted
print(f"The predicted digit is: {predictions[0].argmax()}")
plt.show()
```
